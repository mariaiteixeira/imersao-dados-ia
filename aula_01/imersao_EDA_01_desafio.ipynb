## Valores mínimos e máximos dos dados

```python
dados.head()
```

```python
# Vamos tirar os pontos, pois eles estão com a casa de milhar errada
dados['Votes'] = dados['Votes'].str.replace(',', '')
dados.head()
```

```python
votos = dados['Votes']
```

```python
votos.info()
```

```python
# Os votos estão como "object". Vamos alterar isso
votos = votos.astype('int64')
```

```python
min_votos = votos.min()
```

```python
print(f'O número mínimo de votos é de {min_votos}.')
```

```python
# Acima o número não é tão intuitivo de ler - hora de colocar os pontos nas casas certas de milhar!
formatar_min_votos = '{:,}'.format(min_votos).replace(',', '.')
print(f'O número mínimo de votos para um filme é de {formatar_min_votos}.')
```

```python
max_votos = votos.max()
```

```python
formatar_max_votos = '{:,}'.format(max_votos).replace(',', '.')
print(f'O número máximo de votos para um filme é de {formatar_max_votos}.')
```

# Alterações extras 😸

## Adicionar uma nova coluna para filmes com mais de 3h

```python
dados.head()
```

```python
dados['Filme longo (mais de três horas?)'] = dados['Watch Time'].apply(lambda x: "Sim" if x > 180 else "Não")
dados.head()
```

## Só agora percebi: temos uma coluna desnecessária chamada "Unnamed: 0". Tem que ser removida!

```python
remover_unnamed = dados.drop('Unnamed: 0', axis=1)
```

```python
remover_unnamed.head()
```

```python
dados = remover_unnamed
dados.head()
```

## Nova coluna para os filmes recentes (de 2018 pra 2023) e com ótimas avaliações (8.0 pra cima)

```python
# Nessa base há dados não-numéricos na coluna do ano de lançamento. O regex abaixo irá removê-los
dados['Year of Release'] = dados['Year of Release'].str.replace(r'\D', '', regex=True)
```

```python
dados['Year of Release'] = dados['Year of Release'].astype(float)
```

```python
dados['Alta avaliação e filme recente?'] = (dados['Year of Release'] >= 2018) & (dados['Movie Rating'] >= 8)
dados.tail()
```

```python
# Teste: ver se a nossa coluna recém-adicionada funciona
valor_verdadeiro_nova_coluna = dados[dados['Alta avaliação e filme recente?']]
```
