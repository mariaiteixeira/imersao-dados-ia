# Transformações iniciais dos dados

```python
# Observar os nossos dados
dados.head()
```

```python
# Separar as colunas pelas vírgulas
dados = pd.read_csv('dados_imersao - dados.csv', sep=',')
```

```python
# Após ter separado as colunas, checar como os dados estão
dados.head()
```

```python
# Olhar as informações gerais dos dados
dados.info()
```

```python
# Retirar o símbolo de rúpia da frente dos valores
dados['actual_price'] = dados['actual_price'].str.replace('₹', '')
dados['product_price'] = dados['product_price'].str.replace('₹', '')
```

```python
dados.head()
```

```python
# Agora, trocar as vírgulas dos valores por pontos
dados['actual_price'] = dados['actual_price'].str.replace(',', '.')
dados['product_price'] = dados['product_price'].str.replace(',', '.')
dados.head()
```

```python
# Retirar os pontos
dados['actual_price'] = dados['actual_price'].str.replace(r'\.', '')
dados['product_price'] = dados['product_price'].str.replace(r'\.', '')
```

```python
# Declarar variáveis para as colunas que vamos utilizar
preco_atual = dados['actual_price']
preco_produto = dados['product_price']

# Transformar de "object" para "int64", pois iremos fazer operações com os dados
preco_atual = preco_atual.astype({'actual_price': 'int64'})
preco_produto = preco_produto.astype({'product_price': 'int64'})
```

```python
# Adicionar o ponto no final para separar as cadas decimais e mostrar nossos dados
dados['actual_price'] = dados['actual_price'].str.replace(r'(\d{2}$)', r'.\1')
dados['product_price'] = dados['product_price'].str.replace(r'(\d{2}$)', r'.\1')
dados.head()
```

# Estatísticas descritivas para os preços

```python
# Atualizar os valores nas variáveis e testar
preco_atual = dados['actual_price']
preco_produto = dados['product_price']

preco_atual.head()
```

```python
# Converter para float
preco_atual = dados['actual_price'].astype(float)
preco_produto = dados['product_price'].astype(float)
```

```python
# Média dos preços atuais
preco_atual.mean()
```

```python
# Média dos preços disponíveis online
preco_produto.mean()
```

```python
# Legal. Porém, quantos reais seria o valor? Pois, lembrando, eles estão acima em rúpias. Vamos utilizar uma nova biblioteca
# Primeiro, instalamos ela com o pip install. Após isso, importamos e criamos uma nova instância do CurrencyRates
!pip install forex-python
from forex_python.converter import CurrencyRates
c = CurrencyRates()
```

```python
# Salvar o valor do preço atual em uma variável
media_preco_atual_em_rupia = preco_atual.mean()
```

```python
# Salvar o valor do preço online em uma variável
media_preco_online_em_rupia = preco_produto.mean()
```

```python
# Converter para BRL
valor_preco_atual = c.convert('INR', 'BRL', media_preco_atual_em_rupia)
valor_preco_online = c.convert('INR', 'BRL', media_preco_online_em_rupia)
```

```python
# Vamos imprimir os valores da média!
print(f'A média atual é de ₹{media_preco_atual_em_rupia:.2f} e equivale a R${valor_preco_atual:.2f}.')
print(f'A média online é de ₹{media_preco_online_em_rupia:.2f} e equivale a R${valor_preco_online:.2f}.')
```

```python
# Hora de ver as medianas!
mediana_preco_atual = preco_atual.median()
mediana_preco_online = preco_produto.median()
```

```python
# Converter para BRL as medianas
mediana_atual = c.convert('INR', 'BRL', mediana_preco_atual)
mediana_online = c.convert('INR', 'BRL', mediana_preco_online)
```

```python
# Imprimir os valores da mediana
print(f'A mediana atual é de ₹{mediana_preco_atual:.2f} rúpias, ou seja, R${mediana_atual:.2f}.')
print(f'A mediana online é de ₹{mediana_preco_online:.2f} rúpias, ou seja, R${mediana_online:.2f}.')
```

```python
# Calcular o desvio padrão
desvio_padrao_atual = preco_atual.std()
desvio_padrao_online = preco_produto.std()
```

```python
# Converter para BRL
desvpad_atual = c.convert('INR', 'BRL', desvio_padrao_atual)
desvpad_online = c.convert('INR', 'BRL', desvio_padrao_online)
```

```python
# Imprimir os desvios
print(f'O desvio padrão atual é de ₹{desvio_padrao_atual:.2f} rúpias, ou seja, R${desvpad_atual:.2f}.')
print(f'O desvio padrão online é de ₹{desvio_padrao_online:.2f} rúpias, ou seja, R${desvpad_online:.2f}.')
```

```python
# Calcular o valor mínimo e máximo dos preços atuais e imprimir ambos
valor_min_atual = preco_atual.min()
valor_max_atual = preco_atual.max()
```

```python
# Calcular o valor mínimo e máximo dos preços online e imprimir ambos
valor_min_online = preco_produto.min()
valor_max_online = preco_produto.max()
```

```python
# Converter todos para BRL
min_atual = c.convert('INR', 'BRL', valor_min_atual)
max_atual = c.convert('INR', 'BRL', valor_max_atual)
min_online = c.convert('INR', 'BRL', valor_min_online)
max_online = c.convert('INR', 'BRL', valor_max_online)
```

```python
# Imprimir todos valores!
print(f'Em BRL, o valor mínimo atual é de R${min_atual:.2f}, enquanto o máximo é de R${max_atual:.2f}')
print(f'Já para os preços mostrados online, o valor mínimo é de R${min_online:.2f}, enquanto o máximo é de R${max_online:.2f}')
```
